```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.signal as signal
```

```python
# create sine wave

# 5 second long, sampled at 100 fps
time = np.arange(0,5,1/100)
test_sine = 10*np.sin(2*np.pi*3*time)
plt.plot(test_sine)
```

```python
f = 100
t = 1/100

freq_axis = np.fft.fftfreq(len(test_sine), 1/100)
fft_sine = np.fft.fft(test_sine)

L = len(test_sine) 

Amplitude = abs(fft_sine)/L # normalize by length of signal
half_sided_amplitude = 2*Amplitude[0:round(L/2)-1]
```

```python
plt.plot(freq_axis[0:round(L/2)-1], half_sided_amplitude,label = 'nectary')
plt.xlabel('Frequency (Hz)')
plt.ylabel ('Amplitude')
plt.xticks([0,3,10,20,30,40,50])
plt.xlim(0,10)
```
