```python
import matplotlib.pyplot as plt
plt.rcParams['text.usetex'] = False

import numpy as np
import pandas as pd
import glob
from scipy import signal
```

```python
df = pd.read_pickle('./dataFolder/TrackingData_2022_11_10/ErrorAngle/Bode/allMoths_bode.pickle')
```

```python
Mag = []
Phase = []
m_ID = []
Freq = []

for row, data in df.iterrows():
    mag = data['nect_pilifer_bode_mag']
    phase = data['nect_pilifer_bode_phase']
    freq = data['nect_pilifer_bode_freq']
    mothID = data['mothID']
    print(mothID)
    mothID = [mothID]*len(mag) 
    
    Mag.extend(mag)
    Phase.extend(phase)
    m_ID.extend(mothID)
    Freq.extend(freq)
    
df = pd.DataFrame({'mothID': m_ID,
                  'mag': Mag,
                  'phase': Phase,
                  'freq': Freq})
```

```python
df_mag = df.drop(columns = ['phase'])
df_wide = df_mag.pivot_table(
    index='freq',
    columns='mothID',
    values='mag'
)
```

```python
df_phase = df.drop(columns = ['mag'])
df_phase_wide = df_phase.pivot_table(
    index='freq',
    columns='mothID',
    values='phase'
)
```
