```python
cap = cv2.VideoCapture(path_video + videoname)
video_fps = cap.get(cv2.CAP_PROP_FPS),
total_frames = cap.get(cv2.CAP_PROP_FRAME_COUNT)
height = cap.get(cv2.CAP_PROP_FRAME_HEIGHT)
width = cap.get(cv2.CAP_PROP_FRAME_WIDTH)
print(video_fps, total_frames, height, width)
```

```python
# get the points and make a line for body axis and tangent

body =  df.loc[:, ['body_X', 'body_Y']]
prob_base = df.loc[:, ['prob_base_X', 'prob_base_Y']].values

tangent = pd.read_csv('./dataFolder/ProboscisVector.csv')

p_fit = np.load('./dataFolder/3DpolyfitVariables.npy') #load the coefficients of the polyfit
Y_arr = pd.read_csv('./dataFolder/y_proboscis_ordered', usecols=range(1,7)) #proboscis points in the right order
```

```python
Err_smooth = np.load('./dataFolder/TangentErrorAngle_smoothened.npy')
Err_smooth
```

```python
Y_arr
```

```python
for ff in np.arange(int(total_frames)):
    cap.set(cv2.CAP_PROP_POS_FRAMES, ff)
    ret, frame = cap.read()
    
    # get a point to the proboscis base
    b_x = prob_base[ff][0]
    b_y = prob_base[ff][1]
    
    fig, ax = plt.subplots()
    # zoom into the head region
    plt.xlim([b_x - 150, b_x + 150])
    plt.ylim([b_y + 150, b_y - 150])

    # get points to trace the proboscis reconstruction
    proboscis_y = Y_arr.loc[ff,:].values
    p = np.poly1d(p_fit[ff])
    proboscis_x = p(proboscis_y)
    # create tuple of x,y for the proboscis fit line
    merged_list = np.int32(np.array(tuple(zip(proboscis_x, proboscis_y))))
    
    
    # plot the lines and points
    for pp in np.arange(len(merged_list)-1):
        frame = cv2.line(frame,
                        (merged_list[pp][0], merged_list[pp][1]),
                         (merged_list[pp+1][0], merged_list[pp+1][1]),
                        (0,255,255), 2)
#     frame = cv2.polylines(frame, merged_list, False,(255, 0, 0),2)
#     plt.plot(tangent_x, tangent_y, 'ok-')
    
    frame = cv2.line(frame, 
                     (int(body.iloc[ff,0]), int(body.iloc[ff,1])),
                     (int(prob_base[ff][0]), int(prob_base[ff][1])), 
                     (255, 0, 0), 2)
#     frame = cv2.line(frame, 
#                      (int(tangent_x[ff][0]), int(tangent_y[ff][0])),
#                      (int(tangent_x[ff][1]), int(tangent_y[ff][1])), 
#                      (0, 255, 0), 2)
    ax.imshow(frame)

    ###########################################################################
    new_ax = fig.add_axes([0.4,0.65,0.3,0.2], anchor='NE', zorder=1)
    new_ax.patch.set_facecolor('white')
    new_ax.patch.set_alpha(0.2)
    f = 100
    t = 1/f

    time = np.arange(len(Err_smooth))*t

    plt.plot(time,Err_smooth)
    new_ax.plot(time,Err_smooth*180/np.pi)
    new_ax.plot(ff/f, Err_smooth[ff]*180/np.pi, 'ok', markersize = 4)
    if ff/f < 2:
        new_ax.set_xlim([0,2])
    else:
        new_ax.set_xlim([ff/f - 1, ff/f + 1])
        
    new_ax.text(.75, .75, str(round(Err_smooth[ff]*180/np.pi,2)), c = 'white', 
#                ha='right', va='bottom',
                transform=plt.gca().transAxes)
    
    new_ax.spines['bottom'].set_color('white')
    new_ax.spines['left'].set_color('white')
    new_ax.xaxis.label.set_color('white')
    new_ax.yaxis.label.set_color('white')
    new_ax.tick_params(colors='white')
    new_ax.set_xlabel('Time (ms)', color = 'white')
    new_ax.set_ylabel('Error angle (deg)', color = 'white')
#     ##############################################################################
    
    plt.savefig('./dataFolder/errorSignal/movie/tmp3/' + str(ff).zfill(4) + '.png')
    plt.close()
```
