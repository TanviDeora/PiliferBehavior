```python
import matplotlib.pyplot as plt
plt.rcParams['text.usetex'] = False

import numpy as np
import pandas as pd
import glob
from scipy import signal
```

```python
def half_fft(y, t = None, dt = None, welch=False):
    """
    Compute the one-sided FFT of a signal.

    Parameters:
        y (numpy.ndarray): The input signal in time domain.
        t (numpy.ndarray, optional): The time array corresponding to the signal. If not provided, `dt` must be specified.
        dt (float, optional): The time step (1/sample rate). If not provided, it will be calculated from `t`.

    Returns:
        tuple: A tuple containing:
            - y_fft (numpy.ndarray): The one-sided FFT of the input signal.
            - freq (numpy.ndarray): The corresponding frequency array.

    Notes:
        - If both `t` and `dt` are not provided, the function will return `None` and print an error message.
        - The FFT is normalized using the 'forward' norm, and the amplitudes are doubled for non-DC components.
    """
    
    if dt is None:
        try:
            dt = np.mean(np.diff(t))
        except:
            print('You must either provide an array of times or a dt (1/sample rate).')
            return None
    
    
    if welch:
        # use this to match the frequency resolution to the coherence frequency
        N = len(y)
        nperseg = min(256, N)
        
        # Take the first segment (or a representative chunk)
        y_seg = y[:nperseg] - np.mean(y[:nperseg])

        y_fft = np.fft.rfft(y_seg)
        freq = np.fft.rfftfreq(nperseg, dt)
        y_fft_amp = 2.0 / nperseg * np.abs(y_fft)
    else:
        y_fft = np.fft.fft(y, norm = 'forward')
        freq = np.fft.fftfreq(len(y_fft), dt)

        fidx = freq>=0
        freq=freq[fidx]
        y_fft = y_fft[fidx]
    #     y_fft = np.abs(y_fft)
        y_fft[1:]*=2
        y_fft_amp = np.abs(y_fft)
    
    return y_fft, y_fft_amp, freq
```

```python
filepath= './dataFolder/TrackingData_2022_11_10/thresholded_bodyparts/MV_fixed/'
trackPoints_files = glob.glob(filepath + '*bodyparts.csv')

filepath_another = './dataFolder/TrackingData_2022_11_10/ErrorAngle/MV_fixed/'
pilifer_signal_files = glob.glob(filepath_another + '*errorAngle.csv')
```

```python
cols = ['pilifer_fft','body_fft', 'nectary_fft']

for col in cols:
    uber_df[f'{col}_abs'] = np.abs(uber_df[col])
```

```python
uber_df.head()
```

```python
fft_average = uber_df.drop(columns = ['pilifer_fft', 'body_fft', 'nectary_fft'])
fft_average.head()
fft_average['_t'] = fft_average.groupby(fft_average['mothID']).cumcount()
fft_average = fft_average[['mothID', '_t', 'freq', 'nectary_fft_abs', 'body_fft_abs', 'pilifer_fft_abs']]
stats = fft_average.groupby('_t')[fft_average.columns[2:]].agg(['mean', 'std'])
```

```python
fig, ax = plt.subplots(3,1)
# ax1 = fig.add_axes([0.1,0.1,0.75,0.9])


cols  = ['nectary_fft_abs', 'body_fft_abs', 'pilifer_fft_abs']  
colors = ['black', 'tab:blue', 'tab:orange']

stats = stats.loc[1:, :]
freq = stats[('freq', 'mean')]
# Left axis (first two columns)
for col, c, axes in zip(cols, colors, ax):
    m = stats[(col, 'mean')].values
    s = stats[(col, 'std')].values
    axes.plot(freq, m, color=c, label=col)
    axes.fill_between(freq, m - s, m + s, color=c, alpha = 0.15)
    axes.set_xlim(0,20)

ax[2].set_xlabel('Frequency (Hz)')
fig.text(
    0.04, 0.5,  'Signal Ampltitude',
    va='center',
    rotation='vertical'
)
fig.savefig('./dataFolder/TrackingData_2022_11_10/ErrorAngle/Plots/all_fft.png')
```
